Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as sts
```

```python
# нормальное распределение
mu_values = [2, 3, 1, 1.5 ,4]  # набор значений параметра сдвига 
sigma_values = [0.5, 0.3, 0.6, 0.2, 0.7]  # набор значений параметра масштаба или среденеквадратичное отклонение 
x = np.linspace (0, 4, 100)

plt.figure(figsize=(16, 9))

for i, (mu, sigma) in enumerate(zip(mu_values, sigma_values)):
    sample = sts.norm(loc= mu, scale = sigma)
    sample.rvs(size=20)
    pdf = sample.pdf(x)
    plt.plot(x, pdf, lw=5,label='$\mathcal{mu}'+ '({})$'.format(mu ) + '$\mathcal{sigma}' + '({})$'.format(sigma ) )

plt.legend(fontsize=16)
plt.title('Плотности нормального распределения', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

```python
# экспоненциальнoе распределение
#параметр λ описывает среднее число наступлений события в единицу времени
lam = [0.1, 0.3, 0.4, 0.2 , 1]  # набор значений параметра λ
x1 = np.linspace (0, 5, 100)

plt.figure(figsize=(16, 9))

for lam in lam:
    expon_rv = sts.expon (lam)
    expon_rv.rvs(size = 10)
    pdf1 = expon_rv.pdf(x1)
    plt.plot(x1, pdf1, lw=5,label='$\mathcal{λ}'+ '({})$'.format(lam))

plt.legend(fontsize=16)
plt.title('Плотности экспоненциальнoго распределения', fontsize=10)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.show()
```

```python
# распределение Стьюдента
#параметр n степени свободы (количество случайных величин, взятых для подсчета суммы).
n_values = [4, 2, 3, 10 , 5]  # набор значений параметра n
x2 = np.linspace (-5, 5, 1000)

plt.figure(figsize=(16, 9))

for n in n_values:
    t_rv = sts.t (n)
    t_rv.rvs(size = 10)
    pdf2 = t_rv.pdf(x2)
    plt.plot(x2, pdf2, lw=5,label='$\mathcal{n}'+ '({})$'.format(n))

plt.legend(fontsize=16)
plt.title('Плотности распределения Стьюдента', fontsize=10)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.show()
```
